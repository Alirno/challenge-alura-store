# Ingreso total Estimado por cada tienda

#1. Análisis de facturación



```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
ingreso_tiendas = [float(tiendas['Precio'].sum()) for tiendas in lista_tiendas]

print(ingreso_tiendas)

for i in range(len(ingreso_tiendas)):
  print(f"El ingreso estimado de la tienda {i+1} es de ${ingreso_tiendas[i]:,.0f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

bars = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
y = ingreso_tiendas
x_pos = np.arange(len(bars))

plt.figure(figsize=(9, 5))
plt.bar(x_pos, y, color=['#4300FF', '#0065F8', '#00CAFF', '#00FFDE'], width=0.6,label=bars)
plt.xticks(x_pos, bars)
plt.ylabel('Ingreso')
plt.title('Ingreso Estimado por Tienda')

plt.grid(axis='y', linestyle='--', alpha=0.7)
for i in range(len(y)):
    plt.text(x_pos[i], y[i], f'${y[i]:,.0f}', ha='center', va='bottom')
plt.ylim(min(y) * 0.95, max(y) * 1.05)
plt.legend()
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
cd_productos = {}  # Diccionario general

def cantidad_productos(tienda, i):
    x = list(tienda['Categoría del Producto'])
    conteo = {}  # Diccionario para la tienda actual
    for item in set(x):
        conteo[item] = x.count(item)
    cd_productos[f'tienda{i}'] = conteo  # Se guarda como tienda1, tienda2...

# Recorremos todas las tiendas
for i in range(len(lista_tiendas)):
    cantidad_productos(lista_tiendas[i], i + 1)

df = pd.DataFrame.from_dict(cd_productos, orient='index').T

# Renombrar columnas
df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener etiquetas de categorías (suponiendo que todas las tiendas tienen las mismas)
categorias = list(cd_productos['tienda1'].keys())

# Obtener los valores para cada tienda
tienda1 = list(cd_productos['tienda1'].values())
tienda2 = list(cd_productos['tienda2'].values())
tienda3 = list(cd_productos['tienda3'].values())
tienda4 = list(cd_productos['tienda4'].values())

# Posiciones para las barras
x = np.arange(len(categorias))  # 0, 1, 2, ...

# Tamaño de la figura
plt.figure(figsize=(14, 6))

# Ancho de cada barra
width = 0.1

# Dibujar las barras desplazadas en X
plt.bar(x - 1.5*width, tienda1, width, label='Tienda 1',color='red')
plt.bar(x - 0.5*width, tienda2, width, label='Tienda 2',color='orange')
plt.bar(x + 0.5*width, tienda3, width, label='Tienda 3',color='blue')
plt.bar(x + 1.5*width, tienda4, width, label='Tienda 4',color='green')

# Etiquetas y título
plt.xticks(x, categorias, rotation=45)
plt.ylabel("Cantidad")
plt.title("Cantidad de productos por categoría - Tiendas")
plt.grid(axis='y',linestyle='--', alpha=0.7)

# Leyenda
plt.legend()
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
popularidad_productos = {}  # Diccionario general

def ranking_productos(tienda, i):
    x = list(tienda['Producto'])
    conteo = {}

    # Contar cuántas veces aparece cada producto
    for item in set(x):
        conteo[item] = x.count(item)

    # Ordenar productos por cantidad vendida
    ordenados = sorted(conteo.items(), key=lambda x: x[1], reverse=True)

    # Top 3 más vendidos
    top3_mas = ordenados[:3]

    # Top 3 menos vendidos
    top3_menos = ordenados[-3:]

    # Guardar ambos en el diccionario general
    popularidad_productos[f'tienda{i}'] = {
        'mas_vendidos': top3_mas,
        'menos_vendidos': top3_menos
    }

for i in range(len(lista_tiendas)):
    ranking_productos(lista_tiendas[i], i + 1)
    print(f"\n{'-'*21} Tienda {i+1} {'-'*21}")

    print("Más vendidos:")
    for producto, cantidad in popularidad_productos[f"tienda{i+1}"]["mas_vendidos"]:
        print(f"   - {producto:<35} {cantidad} unidades")

    print("Menos vendidos:")
    for producto, cantidad in popularidad_productos[f"tienda{i+1}"]["menos_vendidos"]:
        print(f"   - {producto:<35} {cantidad} unidades")
    print(f"\n{'-'*52}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def grafico_top(tienda):
    # Unir ambos rankings (más y menos vendidos)
    ranking = popularidad_productos[f'{tienda}']['mas_vendidos'] + popularidad_productos[f'{tienda}']['menos_vendidos']

    # Ordenarlos por la cantidad (de mayor a menor)
    ranking_ordenado = sorted(ranking, key=lambda x: x[1], reverse=True)

    # Extraer productos y cantidades ordenados
    ranking_p = [producto for producto, _ in ranking_ordenado]
    cantidades = [cantidad for _, cantidad in ranking_ordenado]

    x = np.arange(len(ranking_p))

    plt.figure(figsize=(14, 5))
    bars = plt.barh(x, cantidades, color=plt.cm.cool(np.linspace(0, 1, len(ranking_p))), height=0.5)

    plt.yticks(x, ranking_p)
    plt.xlabel("Cantidad")
    plt.title(f"TOP 3 de productos más y menos vendidos - {tienda}")

    # Invertir eje para que el más vendido aparezca arriba
    plt.gca().invert_yaxis()

    # Ajustar los límites del eje x para que se vean mejor las diferencias
    min_val = min(cantidades)
    max_val = max(cantidades)
    plt.xlim(min_val * 0.95, max_val * 1.05)

    # Agregar etiquetas numéricas a las barras
    for i, valor in enumerate(cantidades):
        plt.text(valor+0.5, i, str(valor), va='center',ha='left', fontsize=10)

    plt.tight_layout()
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.show()
```
